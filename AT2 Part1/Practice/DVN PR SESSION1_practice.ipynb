## Groupby

```python
df1.groupby(['Date','Channel'])[['Contribution', 'Impression']].std()
```

```python
df1.groupby(['Date','Channel']).agg({"Contribution": 'mean', "Impression": sum})
```

```python
df1.groupby(['Date','Channel']).agg(mean_contribution=("Contribution",'mean'),
                                    sun_Impression=('Impression', sum),
                                   max_SiteVisit=('Impression', max))
```
