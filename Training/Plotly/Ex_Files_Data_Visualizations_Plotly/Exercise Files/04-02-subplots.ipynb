## Combine Visualizations with Subplots

```
import plotly.express as px
```

### "Facet" plots with `plotly.express`

```
#, facet_col="continent"
#,facet_col_wrap=2

df = px.data.gapminder().query("year == 2007")
fig = px.scatter(df, x="lifeExp", y="gdpPercap", color="country", facet_col="continent",facet_col_wrap=2)
fig.show()
```

### Subplots with `plotly.graph_objects`

```
#limitation of faceting - mixing plot types

import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```
df = px.data.gapminder().query("continent=='Oceania'")
```

```
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "xy"}, {"type": "xy"}]])
```

```
fig.add_trace(go.Bar(x=df[(df['year']==2007) & (df['country']=='New Zealand')]['country'],
                     y=df[(df['year']==2007) & (df['country']=='New Zealand')]['pop']),
              row=1, col=2)
fig.add_trace(go.Bar(x=df[(df['year']==2007) & (df['country']=='Australia')]['country'],
                     y=df[(df['year']==2007) & (df['country']=='Australia')]['pop']),
              row=1, col=2)

fig.add_trace(go.Scatter(x=df[df['country']=='New Zealand']['year'], 
                         y=df[df['country']=='New Zealand']['pop'],
                         mode='lines',
                         name='lines'),
              row=1, col=1)
fig.add_trace(go.Scatter(x=df[df['country']=='Australia']['year'], 
                         y=df[df['country']=='Australia']['pop'],
                         mode='lines',
                         name='lines'),
              row=1, col=1)

fig.update_layout(title='Population of Oceania')
```
