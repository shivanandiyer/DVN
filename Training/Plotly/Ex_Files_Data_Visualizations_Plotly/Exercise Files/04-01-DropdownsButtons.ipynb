## Buttons, Animations & Dropdowns

```
import plotly.express as px
```

```
df = px.data.gapminder()
fig = px.scatter_3d(df[df['continent']=='Europe'], 
                    x='lifeExp',
                    y='pop', 
                    z='gdpPercap',
                    color='country', 
                    animation_frame="year", 
                    animation_group="country",
 )

fig.show()
```

```
import plotly.graph_objects as go

fig = go.Figure(go.Sunburst( 
    labels = ['Earth','Land','Water','Agricultural Land','Non-Agricultural Land','Saline','Freshwater'], 
    parents = ['','Earth','Earth','Land','Land','Water','Water'], 
    values = [1,.29,.71,.10933,.18067,.68515,.02485], 
    branchvalues='total'
)) 
fig
```

```
fig.update_layout(
    updatemenus=[
        go.layout.Updatemenu(
            type="buttons",
            buttons=list([      
                dict(
                    args=[{"marker":""}],
                    label="Original Color",
                    method="update"),
                dict(
                    args=[{"marker.colors": [[1,.29,.71,.10933,.18067,.68515,.02485]]
                           ,"marker.showscale":True
                           ,"marker.colorscale":"Viridis"}],
                    label="Color by % of Earth",
                    method="update")
                ]))])
```
