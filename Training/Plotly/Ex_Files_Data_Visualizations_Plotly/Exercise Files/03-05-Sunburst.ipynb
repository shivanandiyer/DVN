## The Sunburst Chart

```
import plotly.express as px
import pandas as pd
```

```
earth = pd.DataFrame({'labels':['Earth','Land','Water','Agricultural Land','Non-Agricultural Land','Saline','Freshwater'],
                      'parents':['','Earth','Earth','Land','Land','Water','Water'],
                      'values':[1,.29,.71,.10933,.18067,.68515,.02485]})

earth
```

```
fig = px.sunburst(earth,
                  names='labels',
                  parents='parents',
                  values='values',
                  branchvalues='total')

fig.show()
```

### Working with Tidy Data (no data prep required!)

```
df = px.data.gapminder().query("year == 2007")
df.head(3)
```

```
# color_continuous_scale='viridis'

df = px.data.gapminder().query('year==2007')
fig = px.sunburst(df, 
                  path=['continent', 'country'],
                  color='lifeExp', 
                  values='pop')
fig.show()
```
