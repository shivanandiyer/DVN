## Creating a Sankey with Plotly

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/1/10/JIE_Sankey_V5_Fig1.png/800px-JIE_Sankey_V5_Fig1.png">

### To create our Sankey, we'll need to use `graph_objects`

```
import plotly.graph_objects as go
import pandas as pd
```

```
salmon = pd.read_csv('salmon-survival.csv')
```

### Example: Salmon Survival

1. Three species of Salmon are released into a lake (Chinook, Sockeye, and Coho)
2. Of these fish, some will be re-captured and released beyond a dam so they can migrate out to the sea. Those that aren't re-captured either perished or remain in the lake permanently.
3. After travelling out to the open ocean, the fish mature and attempt to return to their home waters to spawn.
4. This process takes a few years, so many will not make the return journey. 

```
salmon
```

```
fig = go.Figure(data=[go.Sankey(
    node = dict(
      label = ['Chinook Released','Coho Released','Sockeye Released','Re-Captured','Not Re-Captured','Returned','Not Re-Captured'],
    ),
    link = dict(
      source = salmon['source'],
      target = salmon['target'],
      value = salmon['value']
  ))])
```

```
fig.update_layout(title_text="Salmon Survival Sankey Diagram", font_size=10)
fig.show()
```
